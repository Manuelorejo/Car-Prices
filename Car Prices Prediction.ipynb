```python
#Binning the Car Companies based on avg prices of each Company.
df['price'] = df['price'].astype('int')
temp = df.copy()
table = temp.groupby(['Company'])['price'].mean()
temp = temp.merge(table.reset_index(), how='left',on='Company')
bins = [0,10000,20000,40000]
df_bin=['Budget','Medium','Highend']
df['carsrange'] = pd.cut(temp['price_y'],bins,right=False,labels=cars_bin)
df.head()
```

After analysis the relevant columns are "wheelbase","carlength",carwidth", "curbweight","enginesize","horsepower", "Carsrange","fueltype","drivewheel","enginetype","cylindernumber","fuelsystem" and "boreratio" columns

```python
cars = df[["price","wheelbase","carlength","carwidth", "curbweight","enginesize","horsepower","fueltype","drivewheel","enginetype","cylindernumber","fuelsystem","boreratio","carsrange"]]
```

```python
sns.pairplot(cars)
```

```python
cars['carsrange']
```

```python
def OHE(x,df):
    temp_df = pd.get_dummies(df[x],drop_first=True)
    df = pd.concat([df,temp_df], axis = 1)
    df.drop(x,axis =1, inplace = True)
    return df
```

```python
cars.info()
```

```python
cars = OHE("fueltype",cars)
cars = OHE("drivewheel",cars)
cars = OHE("enginetype",cars)
cars = OHE("cylindernumber",cars)
cars = OHE("fuelsystem",cars)
cars = OHE("carsrange",cars)
```

The code above replaces the categorical columns in the dataframe with their dummy variables and removes the original column

```python
cars.drop([200,201,202,203,204],inplace = True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
train_data = scaler.fit_transform(cars.drop("price",axis =1 ))
```

```python
X = train_data
y = cars['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred  = model.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
